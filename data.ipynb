```python
cov_states_url = "https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-states.csv"

df = pd.read_csv(cov_states_url)

df["state"] = df["state"].apply(lambda x: us_state_to_abbrev[x])

df.drop(["date", "fips"], axis=1, inplace=True)

df.rename(columns={"state":"State"}, inplace=True)

df = df.groupby("State", as_index=False).last()

df = df.drop(df[
    (df["State"] == "VI") |
    (df["State"] == "UM") | 
    (df["State"] == "PR") |
    (df["State"] == "MP") |
    (df["State"] == "GU") |
    (df["State"] == "AS") | 
    (df["State"] == "DC") 
    ].index)

df.reset_index(inplace=True)
df.drop("index", axis=1, inplace=True)

df[["State", "deaths", "cases"]].to_csv("data/processed/radial/cov_usa_totals.csv", index=False)
```

```python
cov_counties_url = "https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv"

df = pd.read_csv(cov_counties_url)

df["state"] = df["state"].apply(lambda x: us_state_to_abbrev[x])

df.drop(["date", "fips"], axis=1, inplace=True)

df.rename(columns={"state":"State"}, inplace=True)

# df = df.groupby("State", as_index=False).last()
for state, state_abbrev in us_state_to_abbrev.items():
    if state_abbrev not in ["VI", "UM", "PR", "MP", "GU", "AS","DC"]:
        new_df = df[df.State == state_abbrev].groupby("county").last()
        new_df.drop("State", axis=1, inplace=True)
        new_df = new_df.reset_index()
        new_df.rename(columns={"county":"State"}, inplace=True)
        # new_df = df[df.State == "WA"].groupby("county").last()
        # new_df.drop("State", axis=1, inplace=True)

        file_path = f"data/processed/radial/cov_{state_abbrev}_totals.csv"
        new_df[["State", "deaths", "cases"]].to_csv(file_path, index=False)
    # break
```
